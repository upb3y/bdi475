Question 2: find the top 5 mean income state.

```python
df_by_state = df.copy()
df_by_state['Total Household income'] = df_by_state['sum_w']*df_by_state['Mean']
df_by_state.drop(columns=['id','State_Code','Zip_Code','ALand','AWater','Lon','Lat','Mean','Median','Stdev','County','City','Place','Type','Primary','Area_Code'],inplace=True)
df_by_state = df_by_state.groupby(by='State_Name',as_index=False).agg('sum')
df_by_state['mean']=df_by_state['Total Household income']/df_by_state['sum_w']
```

```python
df_by_state.sort_values('mean',ascending=False,inplace=True)
df_by_state.head(5)
```

Communicate my findings:

```python
fig = px.scatter(df, x="ALand",y="Mean", title='relation between land and mean income')
fig.show()
```

```python
fig = px.scatter(df, x="AWater",y="Mean",title='relation between water usage and mean income')
fig.show()
```

```python

fig = px.scatter(df_by_state, x="Total Household income",y="mean",title='relation between total income and mean income')
fig.show()
```

```python
fig = px.pie(df_by_state, values='Total Household income', names='State_Name', title = 'total income of united states')
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')

fig.show()
```

As we can see in the above graphs, the mean income is not related with size of the land and water usage. The US mean income by states largely falls in 50k to 60k. The states with largest total income (california and illinois) doesn't have a highest mean income among the states. DC has the highest average income of 89562.
